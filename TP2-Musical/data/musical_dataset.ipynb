```python
import pandas as pd
# Load the dataset from the Spotify_HuggingFace_Dataset.csv file
df = pd.read_csv("Spotify_HuggingFace_Dataset.csv")
```

```python
# Display the columns of the dataset
df.columns
```

```python
# Show the summary of the DataFrame, including the number of non-null entries and data types
df.info()
```

```python
# Count the occurrences of each genre in the 'track_genre' column
df['track_genre'].value_counts()
```

```python
# Group by 'track_genre' and get the top 10 genres by average popularity
df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False).head(10)
```

```python
# Get the unique genres present in the 'track_genre' column
df['track_genre'].unique()
```

```python
# # Filter the DataFrame to show the 'tempo' of tracks in the 'opera' genre
df[df['track_genre'] == 'opera']['tempo']
```
